```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
advert = pd.read_csv("/content/drive/MyDrive/Copy of Advertising.csv")
```

```python
advert.head()
```

```python
# find corelation 
print(advert["TV"].corr(advert["Sales"]))
```

```python
# anothter

print(advert.TV.corr(advert.Radio))
```

```python
#corelation matrix
print(advert.corr())
```

```python
#show as a Heatmap
dataplot = sns.heatmap(advert.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

```python
#show as a scatterplot

scatter_plot = advert.plot.scatter(x = "TV" , y = "Sales" , color = "g")
plt.show()
```

# Covariance



```python
#covariance co efficient
print(advert["TV"].cov(advert["Newspaper"]))
```

```python
#covariance matrix
print(advert.cov())
```

```python
#fmt = "g" means use genral format for numbers, not scientific notations
dataplot = sns.heatmap(advert.cov(),annot = True )
```

```python
# use fmt = "g"
dataplot = sns.heatmap(advert.cov(),annot = True,fmt = "g")
plt.show()
```
